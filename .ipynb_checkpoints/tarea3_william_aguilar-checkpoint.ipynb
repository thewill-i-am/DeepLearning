```python
im = Image.fromarray(np.array(einstein, dtype = "uint8"),mode='L')
im.resize((256,256))
```

```python
def aplicarFiltro(km, scale):
    k = ImageFilter.Kernel(size=(3, 3),kernel=km, scale=scale)
    im2 = im.filter(k)
    return im2
```

```python
encoder = LabelEncoder()

def transformarOneHot(y, encoder):
    imagenes = encoder.fit_transform(y)
    return imagenes
```

```python
x_test = np.load("imagenesClase/imagenes_x_test.npy")
x_train = np.load("imagenesClase/imagenes_x_train.npy")
y_test = np.load("imagenesClase/imagenes_y_test.npy")

y_train = np.load("imagenesClase/imagenes_y_train.npy")
```

```python
num_labels = len(np.unique(y_train))
print("NÃºmero de etiquetas: ", num_labels,"\n")

```

```python
imgplot = plt.imshow(x_train[1])
plt.show()
```

```python
def preparacionDatos(x):
    image_size = x.shape[1]
    x_prepared = np.reshape(x,[-1, image_size, image_size, 3])
    x_prepared = x_prepared.astype('float32') / 255
    return x_prepared

x_test = preparacionDatos(x_test)
x_train = preparacionDatos(x_train)
```

```python
x_test
```

```python
x_train
```

```python

y_train = transformarOneHot(y_train, encoder)
y_test = transformarOneHot(y_test, encoder)
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
print('Training data shape : ', x_train.shape, y_train.shape)
```

```python
print('Test data shape : ', x_test.shape, y_test.shape)
```

```python
modelo = Sequential()
modelo.add(Conv2D(filters=32, kernel_size=(3),
                       activation='relu',input_shape=(32,32,3)))

modelo.add(MaxPooling2D(2))

modelo.add(Conv2D(filters=32, kernel_size=(3),
                       activation='relu'))

modelo.add(MaxPooling2D(2))

modelo.add(Flatten())

modelo.add(Dropout(0.2))

modelo.add(Dense(units=50, activation="relu"))
modelo.add(Dense(units=10, activation="relu"))

modelo.add(Activation("softmax"))

modelo.compile(loss=keras.losses.categorical_crossentropy,
                    optimizer='adam',
                    metrics=['accuracy'])
modelo.summary()

modelo.fit(x_train, y_train, epochs=50 , batch_size=20, verbose = 0)
```

```python
loss, acc = modelo.evaluate(x_test, y_test, batch_size=16, verbose=0)
```

```python
loss
```

```python
acc
```

```python
pred =  modelo.predict(x_test)
```

```python
pred = np.argmax(pred, axis = -1)
print("Predicciones: ",pred, "\n")
```

Pues como podemos ver de acuerdo a las regresiones presentes no tenemos el mejor modelo predictivo podemos ver que el resultado global es menor al 60%, y el error es alto, pero sin embargo note que para ciertas clases si predice de una forma correcta, en la clase de los animales que son osos por ejemplo me parece que los toma como manzanas, pero en otras clases como bicicletas, peses u abejas realmente predice bien y eso se puede ver en los resultados de la precision por categoria.

```python
image_size = x_train.shape[1]
input_shape = (image_size, image_size, 1)
modelo = Sequential()
modelo.add(Conv2D(filters=64, kernel_size=(3),
                       activation='sigmoid',input_shape=input_shape))

modelo.add(MaxPooling2D(2))

modelo.add(Conv2D(filters=64, kernel_size=(3),
                       activation='sigmoid', input_shape=input_shape))

modelo.add(MaxPooling2D(2))

modelo.add(Flatten())

modelo.add(Dropout(0.2))

modelo.add(Dense(units=50, activation="sigmoid"))
modelo.add(Dense(units=10, activation="sigmoid"))

modelo.add(Activation("softmax"))

modelo.compile(loss=keras.losses.categorical_crossentropy,
                    optimizer='RMSprop',
                    metrics=['accuracy'])
modelo.summary()
```
