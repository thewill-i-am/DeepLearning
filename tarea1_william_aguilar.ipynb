```python
figsize = (6,4)
dpi = 150
def f(x):
    return (3*pow(x,4)) - (16 * pow(x,3)) + (18*pow(x,2))

def df(x):
    return (12*pow(x,3)) - (48 * pow(x,2)) + (36*x)

eta = 0.05
x = [2.0]
y = [f(x[0])]
for i in range(1):
    x.append(x[i] - eta * df(x[i]))
    y.append(f(x[i] - eta * df(x[i])))

dt = pd.DataFrame({"X": x, "Y": y})
print(dt)
t1 = np.arange(-5.0, 5.0, 0.1)
fig, ax = plt.subplots(1,1, figsize = figsize, dpi = dpi)
ax.set_xlim(-5, 5)
ax.set_ylim(-45, 45)
ax.plot(t1, f(t1))
ax.plot(x,y,'ro')
```

Ejercicio 9


![alt text](ejercicio9.jpeg "Title")
![alt text](ejercicio9-1.jpeg "Title")
![alt text](ejercicio9-2.jpeg "Title")

```python
def f(x,y):
    return x*x + y*y - 2*x - 6*y + 14


def gradiente_f(x,y):
    return np.array([2*x-2,2*y-6])


x = -4
y = 4
eta = 0.3
xi = np.array([x,y])

x = []
y = []
z = []

for i in range(4):
    xi = xi - eta * gradiente_f(xi[0],xi[1])
    x.append(xi[0])
    y.append(xi[1])
    z.append(f(xi[0],xi[1]))


dataF = pd.DataFrame({"X":x, "Y": y, "Z": z})
#Imprimimos en consola solamente 10
print(dataF.head(n = 10))


#Grfico
fig = px.scatter_3d(dataF, x='X', y='Y', z='Z',color_discrete_sequence=["red"])

xdata = np.arange(-5, 5, 0.1)
ydata = np.arange(-5, 5, 0.1)
X,Y = np.meshgrid(xdata,ydata)
Z = f(X,Y)
fig.add_trace(go.Surface(
    x = X,
    y = Y,
    z = Z,
    opacity = .7, showscale = False,
    colorscale='Viridis'
))

fig.show()
```

![alt text](ejercicio10-1.jpeg "Title")

```python

def f(x,y):
    return pow(x,4) + pow(y,4) - 4*x*y + 1


def gradiente_f(x,y):
    return np.array([((4*pow(x,3))-(4*y)),  ((4*pow(y,3))-(4*x))])


x = -4
y = 4
eta = 0.0009
xi = np.array([x,y])

x = []
y = []
z = []

for i in range(4):
    xi = xi - eta * gradiente_f(xi[0],xi[1])
    x.append(xi[0])
    y.append(xi[1])
    z.append(f(xi[0],xi[1]))


dataF = pd.DataFrame({"X":x, "Y": y, "Z": z})
#Imprimimos en consola solamente 10
print(dataF.head(n = 10))


#Grfico
fig = px.scatter_3d(dataF, x='X', y='Y', z='Z',color_discrete_sequence=["red"])

xdata = np.arange(-5, 5, 0.1)
ydata = np.arange(-5, 5, 0.1)
X,Y = np.meshgrid(xdata,ydata)
Z = f(X,Y)
fig.add_trace(go.Surface(
    x = X,
    y = Y,
    z = Z,
    opacity = .7, showscale = False,
    colorscale='Viridis'
))

fig.show()
```
